# compute cost

```python
def compute_cost(x,y,w,b):
    length=len(x)
    cost=0
    for m in range(length):
        cost+=(w*x[m]+b-y[m])**2
    return cost/(2*m)
```

# grad descend

```python
epoch=20
alpha=0.0001
initial_w=0
initial_b=0 
```

```python
def grad_descent(epoch,alpha,w,b,x,y):
    cost=[]
    
    for num_iters in range(epoch):
        cost.append(compute_cost(x,y,w,b))
        w,b=step_grad_desc(w,b,alpha,x,y)
    return w,b,cost
```

```python
def step_grad_desc(w,b,alpha,x,y):
    length=len(x)
    delta_w=0
    delta_b=0
    for i in range(length):
        delta_w+=(w*x[i]+b-y[i])*x[i]
        delta_b+=w*x[i]+b-y[i]
    w=w-alpha*delta_w/length
    b=b-alpha*delta_b/length
    return w,b
```

```python
w,b,cost=grad_descent(epoch,alpha,initial_w,initial_b,x,y)
```
